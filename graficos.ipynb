```python
import pandas as pd

# Ruta del archivo TSV
path = 'C:/Users/franc/OneDrive/Escritorio/logt1/experiments_data/data.tsv'

```

```python
# Leer el archivo TSV y convertirlo a un DataFrame
data = pd.read_csv(path, sep='\t')

# Mostrar las primeras filas del DataFrame
print(data)
```

```python
import matplotlib.pyplot as plt

# Filtrar los datos por cada valor único de N
unique_N_values = data['N'].unique()

# Crear un gráfico para cada valor de N
for N_value in unique_N_values:
    subset = data[data['N'] == N_value]
    plt.figure()
    plt.plot(subset['C_MAX'], subset['costo_promedio'], marker='o')
    plt.title(f'Costo Promedio vs C_MAX para N={N_value}')
    plt.xlabel('C_MAX')
    plt.ylabel('Costo Promedio')
    plt.grid(True)
    plt.show()
```

```python
# Filtrar los datos por cada valor único de N
unique_N_values = data['N'].unique()

# Crear un gráfico para cada valor de N
for N_value in unique_N_values:
    subset = data[data['N'] == N_value]
    plt.figure()
    plt.plot(subset['C_MAX'], subset['io_count'], marker='o')
    plt.title(f'io_count vs C_MAX para N={N_value}')
    plt.xlabel('C_MAX')
    plt.ylabel('io_count')
    plt.grid(True)
    plt.show()
```

```python
# Filtrar los datos por cada valor único de N
unique_N_values = data['N'].unique()

# Crear un gráfico para cada valor de N
for N_value in unique_N_values:
    subset = data[data['N'] == N_value]
    plt.figure()
    plt.plot(subset['C_MAX'], subset['porcentaje_llenado'], marker='o')
    plt.title(f'Porcentaje llenado vs C_MAX para N={N_value}')
    plt.xlabel('C_MAX')
    plt.ylabel('Porcentaje llenado')
    plt.grid(True)
    plt.show()
```

```python
# Gráfico de líneas para costo_promedio vs N
plt.figure(figsize=(10, 6))
for C_MAX in data['C_MAX'].unique():
    subset = data[data['C_MAX'] == C_MAX]
    plt.plot(subset['N'], subset['costo_promedio'], marker='o', linestyle='-', label=f'C_MAX={C_MAX}')
plt.xlabel('N')
plt.ylabel('Costo Promedio')
plt.title('Costo Promedio vs N para diferentes C_MAX')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gráfico de dispersión para io_count vs N
plt.figure(figsize=(10, 6))
for C_MAX in data['C_MAX'].unique():
    subset = data[data['C_MAX'] == C_MAX]
    plt.scatter(subset['N'], subset['io_count'], label=f'C_MAX={C_MAX}')
plt.xlabel('N')
plt.ylabel('IO Count')
plt.title('IO Count vs N para diferentes C_MAX')
plt.legend()
plt.grid(True)
plt.show()
```
