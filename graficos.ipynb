```python
import pandas as pd

# Ruta del archivo TSV
path = 'C:/Users/franc/OneDrive/Escritorio/logt1/experiments_data/data.tsv'

```

```python
# Leer el archivo TSV y convertirlo a un DataFrame
data = pd.read_csv(path, sep='\t')

# Mostrar las primeras filas del DataFrame
print(data)
```

```python
import matplotlib.pyplot as plt

# Filtrar los datos por cada valor nico de N
unique_N_values = data['N'].unique()

# Crear un grfico para cada valor de N
for N_value in unique_N_values:
    subset = data[data['N'] == N_value]
    plt.figure()
    plt.plot(subset['C_MAX'], subset['costo_promedio'], marker='o')
    plt.title(f'Costo Promedio vs C_MAX para N={N_value}')
    plt.xlabel('C_MAX')
    plt.ylabel('Costo Promedio')
    plt.grid(True)
    plt.show()
```
